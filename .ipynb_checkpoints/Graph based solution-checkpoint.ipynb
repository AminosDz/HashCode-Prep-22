```python
from reader import read
from Pizza import Pizza
import networkx as nx
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
fname = "d_difficult.in.txt"

clients, ingredients = read(f'data/{fname}')

nb_clients, nb_ing = len(clients), len(ingredients)
CLI = range(nb_clients)
ING = range(nb_ing)
```

```python
ing_like_clients = {}
ing_dislike_clients = {}
for j, c in enumerate(clients):
    for ing in c[0]:
        ing_like_clients.setdefault(ing, []).append(j)
    for ing in c[1]:
        ing_dislike_clients.setdefault(ing, []).append(j)

conflicted_clients = {}
for ing in ing_like_clients:
    for c in ing_like_clients[ing]:
        for other in ing_dislike_clients.get(ing, []):
            conflicted_clients.setdefault(c,set()).add(other)

for ing in ing_dislike_clients:
    for c in ing_dislike_clients[ing]:
        for other in ing_like_clients.get(ing, []):
            conflicted_clients.setdefault(c,set()).add(other)
```

```python
# Decomposition
G = nx.Graph()

all_clients = set(CLI)
G.add_nodes_from([f'C{j}' for j in CLI])

for c in conflicted_clients:
    unconflicted_clients_c = set.difference(all_clients, set(conflicted_clients[c]))
    for other in unconflicted_clients_c:
        if c != other:
            G.add_edge(f'C{c}', f'C{other}')

largest_cc = list(nx.find_cliques(G))
```
